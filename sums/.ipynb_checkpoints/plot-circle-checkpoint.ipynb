```python
# from https://stackoverflow.com/questions/32092899/plot-equation-showing-a-circle
# based on http://matplotlib.org/examples/pylab_examples/contour_demo.html


import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1.0, 1.0, 100)
y = np.linspace(-1.0, 1.0, 100)
X, Y = np.meshgrid(x,y)
F = X**2 + Y**2 - 0.6
plt.contour(X,Y,F,[0])
plt.show()
```
